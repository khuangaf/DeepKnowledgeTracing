```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import sklearn 
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import csr_matrix
```

```python
df= pd.read_csv('data/ASSISTmentsProblems.csv')
df 
```

```python
def parse_html(row):
    soup = BeautifulSoup(row['body'], 'html.parser')
    row['parsed'] = soup.get_text()
    return row
```

```python
df.body = df.body.astype('string')
df = df.apply(parse_html, axis=1)
df
```

```python
vectorizer = TfidfVectorizer()
skill_vector = vectorizer.fit_transform(df['parsed'])
skill_vector
```

```python
denseVec = csr_matrix.todense(skill_vector)
```
