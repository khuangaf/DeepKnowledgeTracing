```python
import pandas as pd
import numpy as np
import re
from tqdm import *
```

```python
data_path = 'data/algebra_2005_2006_test.txt'
temp_path ='data/kdd_train_0506_alg_test_temp.csv'
path = 'data/kdd_train_0506_alg_test.csv'
output_name= 'data/algebra_2005_2006_test.csv'
```

```python
with open(data_path, 'r') as f:
    headers = f.readline()

headers = re.split('\t',headers[:-1])
headers
```

```python
df = pd.DataFrame(columns=headers)
lines = []
with open(data_path, 'r') as f:
    for i, line in enumerate(tqdm(f)):
        if i == 0: continue
        lines.append(re.split('\t',line[:-1]))
# re.split('\t',a)
df = pd.DataFrame(lines, columns=headers)
df
```

```python
student_ids = set(list(df['Anon Student Id']))
len(student_ids)
len(set(list(df['Problem Name'])))
```

<b>Row</b>: the row number Update (04-20-2010): for challenge data sets, the row number in each file (train, test, and submission) is no longer taken from the original data set file. Instead, rows are renumbered within each file. So instead of 1…n rows for the training file and n+1..m rows for the test/submission file, it is now 1…n for the training file and 1…n for the test/submission file. <br>
<b>Anon Student Id</b>: unique, anonymous identifier for a student<br>
<b>Problem Hierarchy</b>: the hierarchy of curriculum levels containing the problem.<br>
<b>Problem Name</b>: unique identifier for a problem<br>
<b>Problem View</b>: the total number of times the student encountered the problem so far.<br>
<b>Step Name</b>: each problem consists of one or more steps (e.g., “find the area of rectangle ABCD” or “divide both sides of the equation by x”). The step name is unique within each problem, but there may be collisions between different problems, so the only unique identifier for a step is the pair of problem_name and step_name.<br>
<b>Step Start Time</b>: the starting time of the step. Can be null.<br>
<b>First Transaction Time</b>: the time of the first transaction toward the step.<br>
<b>Correct Transaction Time</b>: the time of the correct attempt toward the step, if there was one.<br>
<b>Step End Time</b>: the time of the last transaction toward the step.<br>
<b>Step Duration (sec)</b>: the elapsed time of the step in seconds, calculated by adding all of the durations for transactions that were attributed to the step. Can be null (if step start time is null).<br>
<b>Correct Step Duration (sec)</b>: the step duration if the first attempt for the step was correct.<br>
<b>Error Step Duration (sec)</b>: the step duration if the first attempt for the step was an error (incorrect attempt or hint request).<br>
<b>Correct First Attempt</b>: the tutor’s evaluation of the student’s first attempt on the step - 1 if correct, 0 if an error.<br>
<b>Incorrects</b>: total number of incorrect attempts by the student on the step.<br>
<b>Hints</b>: total number of hints requested by the student for the step.<br>
<b>Corrects</b>: total correct attempts by the student for the step. (Only increases if the step is encountered more than once.)<br>
<b>KC(KC Model Name)</b>: the identified skills that are used in a problem, where available. A step can have multiple KCs assigned to it. Multiple KCs for a step are separated by ~~ (two tildes). Since opportunity describes practice by knowledge component, the corresponding opportunities are similarly separated by ~~.
Opportunity(KC Model Name): a count that increases by one each time the student encounters a step with the listed knowledge component. Steps with multiple KCs will have multiple opportunity numbers separated by ~~.
Additional KC models, which exist for the challenge data sets, will appear as additional pairs of columns (KC and Opportunity columns for each model).<br>

```python
#re.match(r'(ftp|http)://.*\.(jpg|png)$', s)
def extractSkill(row):
    skill = row['KC(Default)']
    start= 11
    end = skill.find(';')
    row['KC(Default)'] = skill[start:end]
    return row
    
```

```python
df['KC(Default)'] = df.apply(extractSkill, axis=1)
```

```python
df['skill'] = df['KC(Default)']
df
```

```python
interestedDf=df[['Row','Anon Student Id', 'Correct First Attempt','skill']]
```

```python
skills = list(set(df['skill']))
```

```python
def skillToIndex(row):
    skill = row['skill']
    return skills.index(skill)
```

```python
interestedDf['skill'] = interestedDf.apply(skillToIndex, axis=1)
interestedDf
```

```python
interestedDf.to_csv(output_name, index=False)
```

```python
skill_seqs = []
correct_seqs = []
seq_nums = []
for id in student_ids:
    studentDf = interestedDf.loc[interestedDf['Anon Student Id'] == id, ['Correct First Attempt','skill']]
    seq_num = studentDf.shape[0]
    skill_seq = studentDf['skill']
    correct_seq = studentDf['Correct First Attempt']
    seq_nums.append(seq_num)
    skill_seqs.append(skill_seq)
    correct_seqs.append(correct_seq)
```

```python

for i in range(len(skill_seqs)):
    with open(temp_path,'a') as f_handle:
        sequenceLength = seq_nums[i]
        correct = np.array(correct_seqs[i])
        skill = np.array(skill_seqs[i])
        np.savetxt(f_handle, np.array([sequenceLength]), fmt='%i', newline = '\n')
        np.savetxt(f_handle,skill.astype(int), fmt='%i',newline = ',')
        np.savetxt(f_handle,np.array(['\n']),fmt='%s',newline='')
        np.savetxt(f_handle,correct.astype(int), fmt='%i',newline = ',')
        np.savetxt(f_handle,np.array(['\n']),fmt='%s',newline='')
```

```python
with open(temp_path, 'r') as r, open(path, 'w') as w:    
    for num, line in enumerate(r):
        if num %3!= 0:  
            
            newline = line[:-2] + "\n" if "\n" in line else line[:-1]
        else:
            newline = line               
        w.write(newline)
```
